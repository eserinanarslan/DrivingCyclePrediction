```python
# Load Train dataset
test_1 = pd.read_csv("../data/SurusTesti1.csv", delimiter = ";")
test_2 = pd.read_csv("../data/SurusTesti2.csv", delimiter = ";")
```

```python
# Load cycle_df dataset
cycle_df = pd.read_csv("../data/cycle_df.csv")
cycle_df.head(5).append(cycle_df.tail(5))
```

```python
test_1.head(5).append(test_1.tail(5))
```

```python
test_2.head(5).append(test_2.tail(5))
```

```python
test_1 = test_1.drop({'Phase', 'TotalTime'}, axis=1)
```

```python
test_2 = test_2.drop({'Phase', 'TotalTime'}, axis=1)
```

```python
period = [5, 10, 30, 60]
process = ['Vel', 'Acc']
```

```python
for per in period:
    for pro in process:
        col = str(per) + "_sec_" + str(pro)+ "_mean"
        test_1[col] = test_1[pro].expanding(min_periods=per).mean().round(5)
        test_2[col] = test_2[pro].expanding(min_periods=per).mean().round(5)
        cycle_df[col] = cycle_df[pro].expanding(min_periods=per).mean().round(5)
```

```python
period_2 = [30, 60]
process_2 = ['Acc']
for per in period_2:
    for pro in process_2:
        col_min = str(per) + "_sec_" + str(pro)+ "_min"
        col_max = str(per) + "_sec_" + str(pro)+ "_max"

        test_1[col_min] = test_1[process_2].expanding(min_periods=per).min().round(5)
        test_1[col_max] = test_1[process_2].expanding(min_periods=per).max().round(5)

        test_2[col_min] = test_2[process_2].expanding(min_periods=per).min().round(5)
        test_2[col_max] = test_2[process_2].expanding(min_periods=per).max().round(5)
        
        cycle_df[col_min] = cycle_df[process_2].expanding(min_periods=per).min().round(5)
        cycle_df[col_max] = cycle_df[process_2].expanding(min_periods=per).max().round(5)
```
