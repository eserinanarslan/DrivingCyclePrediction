# Preprocess Steps

```python
#Set parameters to see all data
pd.set_option('display.max_rows', 150)
pd.set_option('display.max_columns', 50)
pd.set_option('display.width', 1000)
```

```python
# Load Train dataset
test_1 = pd.read_csv("../data/SurusTesti1.csv", delimiter = ";")
test_2 = pd.read_csv("../data/SurusTesti2.csv", delimiter = ";")
```

```python
# Load cycle_df dataset
cycle_df = pd.read_csv("../data/cycle_df_for_main_cycles.csv")
cycle_df.head(5).append(cycle_df.tail(5))
```

```python
test_1.head(5).append(test_1.tail(5))
```

```python
test_2.head(5).append(test_2.tail(5))
```

```python
test_1.shape, test_2.shape, cycle_df.shape
```

```python
test_1 = test_1.drop({'Phase', 'TotalTime'}, axis=1)
```

```python
test_2 = test_2.drop({'Phase', 'TotalTime'}, axis=1)
```

```python
period = [5, 10, 30, 60]
process = ['Vel', 'Acc']
```

```python
for per in period:
    for pro in process:
        col = str(per) + "_sec_" + str(pro)+ "_mean"
        test_1[col] = test_1[pro].expanding(min_periods=per).mean().round(5)
        test_2[col] = test_2[pro].expanding(min_periods=per).mean().round(5)
        cycle_df[col] = cycle_df[pro].expanding(min_periods=per).mean().round(5)
```

```python
period_2 = [30, 60]
process_2 = ['Acc']
for per in period_2:
    for pro in process_2:
        col_min = str(per) + "_sec_" + str(pro)+ "_min"
        col_max = str(per) + "_sec_" + str(pro)+ "_max"

        test_1[col_min] = test_1[process_2].expanding(min_periods=per).min().round(5)
        test_1[col_max] = test_1[process_2].expanding(min_periods=per).max().round(5)

        test_2[col_min] = test_2[process_2].expanding(min_periods=per).min().round(5)
        test_2[col_max] = test_2[process_2].expanding(min_periods=per).max().round(5)
        
        cycle_df[col_min] = cycle_df[process_2].expanding(min_periods=per).min().round(5)
        cycle_df[col_max] = cycle_df[process_2].expanding(min_periods=per).max().round(5)
```

```python
period_2 = 60
process_2 = 'Vel'

# dummy function to count zeros
count_func = lambda x: (x==0).sum()

iddle_count = str(period_2) + "_sec_" + str(process_2)+ "_iddle_count"
print("******* " + iddle_count + " *******")

test_1[iddle_count] = 0
test_1[iddle_count] = test_1['Vel'].rolling(60).apply(count_func)

test_2[iddle_count] = 0
test_2[iddle_count] = test_2['Vel'].rolling(60).apply(count_func)

cycle_df[iddle_count] = 0
cycle_df[iddle_count] = cycle_df['Vel'].rolling(60).apply(count_func)
```

# Prediction Steps

```python
phase_mapping = {
    "artemis_motor130" : "10",
    "artemis_road" : "30",
    "artemis_urban" : "50",
    "ftp72" : "70",
    "ftp_us06" : "90",
    "hwfet" : "110",
    "nycc" : "130",
    "rts95" : "150",
    "wltp_class3b" : "170",
}
```

```python
invertion_phase_mapping = {
    "10" : "artemis_motor130",
    "30" : "artemis_road",
    "50" : "artemis_urban",
    "70" : "ftp72",
    "90" : "ftp_us06",
    "110" : "hwfet",
    "130" : "nycc",
    "150" : "rts95",
    "170" : "wltp_class3b",
}
```

```python
cycle_df.head()
```

```python
cycle_df.shape
```

```python
def convert_phase(df, column_name, phase_dict):
    df[column_name] = df[column_name].replace(phase_dict)
    return df
```

```python
cycle_df = convert_phase(cycle_df, "Phase", phase_mapping)
```

```python
cycle_df = cycle_df.drop({'TotalTime'}, axis=1)
```

```python
# load the model from disk
filename = '../models/KNC_model_for_main_cycles.sav'
loaded_model = pickle.load(open(filename, 'rb'))
```

```python
best_KNC = loaded_model.fit(X, y)
phase_predictor_1 = best_KNC.predict(test_1)
phase_predictor_2 = best_KNC.predict(test_2)

print(phase_predictor_1)
```
