```python
period = [5, 10, 30, 60]
process = ['Vel', 'Acc']
```

```python
artemis_motor130.head()
```

```python
for per in period:
    for pro in process:
        col = str(per) + "_sec_" + str(pro)+ "_mean"
        artemis_motor130[col] = artemis_motor130[pro].expanding(min_periods=per).mean().round(5)
        artemis_road[col] = artemis_road[pro].expanding(min_periods=per).mean().round(5)
        artemis_urban[col] = artemis_urban[pro].expanding(min_periods=per).mean().round(5)
        ftp72[col] = ftp72[pro].expanding(min_periods=per).mean().round(5)
        ftp_us06[col] = ftp_us06[pro].expanding(min_periods=per).mean().round(5)
        hwfet[col] = hwfet[pro].expanding(min_periods=per).mean().round(5)
        nycc[col] = nycc[pro].expanding(min_periods=per).mean().round(5)
        rts95[col] = rts95[pro].expanding(min_periods=per).mean().round(5)
        wltp_class3b[col] = wltp_class3b[pro].expanding(min_periods=per).mean().round(5)
```

```python
period_2 = [30, 60]
process_2 = ['Acc']
for per in period_2:
    for pro in process_2:
        col_min = str(per) + "_sec_" + str(pro)+ "_min"
        col_max = str(per) + "_sec_" + str(pro)+ "_max"

        artemis_motor130[col_min] = artemis_motor130[process_2].expanding(min_periods=per).min().round(5)
        artemis_motor130[col_max] = artemis_motor130[process_2].expanding(min_periods=per).max().round(5)

        artemis_road[col_min] = artemis_road[process_2].expanding(min_periods=per).min().round(5)
        artemis_road[col_max] = artemis_road[process_2].expanding(min_periods=per).max().round(5)

        artemis_urban[col_min] = artemis_urban[process_2].expanding(min_periods=per).min().round(5)
        artemis_urban[col_max] = artemis_urban[process_2].expanding(min_periods=per).max().round(5)

        ftp72[col_min] = ftp72[process_2].expanding(min_periods=per).min().round(5)
        ftp72[col_max] = ftp72[process_2].expanding(min_periods=per).max().round(5)

        ftp_us06[col_min] = ftp_us06[process_2].expanding(min_periods=per).min().round(5)
        ftp_us06[col_max] = ftp_us06[process_2].expanding(min_periods=per).max().round(5)

        hwfet[col_min] = hwfet[process_2].expanding(min_periods=per).min().round(5)
        hwfet[col_max] = hwfet[process_2].expanding(min_periods=per).max().round(5)

        nycc[col_min] = nycc[process_2].expanding(min_periods=per).min().round(5)
        nycc[col_max] = nycc[process_2].expanding(min_periods=per).max().round(5)

        rts95[col_min] = rts95[process_2].expanding(min_periods=per).min().round(5)
        rts95[col_max] = rts95[process_2].expanding(min_periods=per).max().round(5)

        wltp_class3b[col_min] = wltp_class3b[process_2].expanding(min_periods=per).min().round(5)
        wltp_class3b[col_max] = wltp_class3b[process_2].expanding(min_periods=per).max().round(5)
```
